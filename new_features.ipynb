```python
import pandas as pd
import numpy as np
import nltk
import emoji
from nltk.tokenize import word_tokenize,sent_tokenize
import re
```

```python
df=pd.read_csv('withStockMovement_2.csv')
```

```python
df.head()
```

```python
df1=df.head(100)
```

```python
def EmojiPercentage(row):
    text=row.original_text
    chars=list(text)
    count=0
    for char in chars:
        if char in emoji.UNICODE_EMOJI['en']:
            count+=1
    return count/row.length
```

```python
df1['emoji/length'] = df1.apply(EmojiPercentage, axis = 1)
```

```python
df1
```

```python
def count_capital_words(row):
    text=row.original_text
    return sum(map(str.isupper,text.split()))
```

```python
df1['capitalWordCount']=df1.apply(count_capital_words, axis = 1)
```

```python
df1
```

```python
def count_sent(row):
    text=row.original_text
    return len(nltk.sent_tokenize(text))
```

```python
df1['sent_count']=df1.apply(count_sent, axis = 1)
```

```python
df1
```

```python
def count_htags(row):
    text=row.original_text
    x = re.findall(r'(#w[A-Za-z0-9]*)', text)
    return len(x) 
```

```python
df1['hashtag_count']=df1.apply(count_htags, axis = 1)
```

```python
def hasLink(row):
    text=row.original_text
    if 'http' in text:
        return 1
    return 0
    
```

```python
df1['has_link']=df1.apply(hasLink, axis = 1)
```

```python
def stock_slang_richness(row):
    new_words=open('new_words_list.txt')
    new_words=new_words.read()
    new_words=new_words.split('\n')
    text=row.original_text
    total=0
    for word in new_words:
        if word in text:
            total+=text.count(word)
    return total/row.length
```

```python
df1['stock_slang_richness']=df1.apply(stock_slang_richness, axis = 1)
```
